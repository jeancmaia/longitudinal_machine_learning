##### A periodogram

```
In the area of time series called spectral analysis, we view a time series as a sum of cosine waves with varying amplitudes and frequencies. One goal of an analysis is to identify the important frequencies (or periods) in the observed series. A starting tool for doing this is the periodogram. The periodogram graphs a measure of the relative importance of possible frequency values that might explain the oscillation pattern of the observed data.
```
https://online.stat.psu.edu/stat510/lesson/6/6.1


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('DailyDelhiClimateData/DailyDelhiClimateTrain.csv')
data['date'] = pd.to_datetime(data['date'])

sns.lineplot(x='date', y='humidity', data=data)
_ = plt.xticks(rotation=45)
````

````python
from scipy.signal import periodogram

freq, pxx = periodogram(data['humidity'])

sns.lineplot(pxx[0:200])
````

#### Testing for the presence of Hidden Periodicities
----------------

Some tests can be approached on top of periodogram, where the null hypothesis $H_0$ that the data is generated by a Gaussian white noise sequence, where H1 is imposed by deterministic periodic component.

(i) Testing for the Presence of a Sinusoid with Specified Frequency. 

$$ X_t = \mu + A cos \omega t + B sin \omega t + \epsilon_t $$

(ii) Testing for the Presence of a Non-Sinusoidal Periodic Component with a specified integer-valued period, p < n.    

(iii) Testing for Hidden Periodicities of Unspecified Frequency.

#### Spectral Density

Harbouring the spectral philosophy, where a series can be represented as sum of cosine and sine waves, oscillating at some harmonical frequency. The `spectral density` is a frequency domain representation of a time series, related to the autocovariance time domain representation.   

Let $\gamma(h)$ be the autocovariance function and $f(w)$, the spectral density of the same process. $\gamma(h)$ and $f(w)$ are `Fourier transform pairs`.  

````python
from scipy.signal import welch

freq, pxx = welch(data['humidity'])

sns.lineplot(pxx[0:40])
````
